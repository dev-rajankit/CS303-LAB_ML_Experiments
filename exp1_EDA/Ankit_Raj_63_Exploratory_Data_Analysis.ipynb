# [Objective 03](#pandas-visualizations) Describe and discriminate between basic data types

## Overview

One of the cornerstones of Exploratory Data Analysis (EDA) is being able to identify variable types.  We will need different statistical methods to display and describe each of these different types of data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Step 1: Clean & Prepare Data

```python
# Load dataset
iris_df = pd.read_csv("iris_dirty.csv")

# Inspect structure
iris_df.info()
```

```python
# Drop unnecessary identifier column
iris_df = iris_df.drop(columns=["Unnamed: 0"])

# Standardize categorical labels
iris_df["Species"] = iris_df["Species"].str.strip().str.lower()

# Confirm counts of categories
iris_df["Species"].value_counts()
```

Step 2: Understand Variable Types

```python
# Separate numeric and categorical columns
num_cols = iris_df.select_dtypes(include=["float64", "int64"]).columns
cat_cols = iris_df.select_dtypes(include=["object"]).columns

print("Numeric Columns:", list(num_cols))
print("Categorical Columns:", list(cat_cols))

print("\n All four flower measurements (sepal & petal length/width) are continuous numeric variables.")
print(" 'Species' is a categorical (nominal) variable with 3 categories (setosa, versicolor, virginica).")
```

Step 3: Univariate Distributions

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))

for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i)
    sns.histplot(iris_df[col], bins=20, kde=True, color="cornflowerblue")
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

Observations:

- Petal length & width show multimodal patterns, separating species well.

- Sepal width appears normally distributed but has missing values.

- Sepal length shows more overlap across species.

Step 4: Categorical-Numerical Relationship

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Species", y="Sepal.Length", data=iris_df, hue="Species", palette="Set2", legend=False)
plt.title("Sepal Length Distribution by Species")
plt.show()
```

```python
plt.figure(figsize=(12,8))

for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x="Species", y=col, data=iris_df, hue="Species", palette="pastel", legend=False)
    plt.title(f"{col} by Species")

plt.tight_layout()
plt.show()
```

Step 6: Pairwise Relationships

```python
sns.pairplot(iris_df, hue="Species", vars=num_cols, diag_kind="kde", palette="husl")
plt.suptitle("Pairwise Scatterplots of Features", y=1.02)
plt.show()
```

Step 7: Correlation Heatmap

```python
plt.figure(figsize=(8,6))
sns.heatmap(iris_df[num_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```
